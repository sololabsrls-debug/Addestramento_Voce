```python
# CELLA 4: Import e verifica
import torch
from TTS.api import TTS
import soundfile as sf

print("üéÆ GPU Check:")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"VRAM: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

print("\n‚úÖ Import completato!")
```

```python
# CELLA 5: Carica modello XTTS v2
print("üîÑ Caricamento XTTS v2...")
device = "cuda" if torch.cuda.is_available() else "cpu"
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)
print("‚úÖ XTTS v2 pronto!")
```

```python
# CELLA 6: Test sintesi italiana
print("üß™ Test sintesi italiana...")

test_text = "Benvenuto! Questo √® un test di sintesi vocale italiana con XTTS versione 2."

tts.tts_to_file(
    text=test_text,
    file_path="/content/test_output.wav",
    language="it"
)

print("‚úÖ Test completato!")
print("üìÅ File: /content/test_output.wav")
```
