# üé§ XTTS v2 su Google Colab - Python 3.11

‚ö†Ô∏è **Soluzione per Python 3.12**  
Google Colab usa Python 3.12, ma TTS richiede Python ‚â§3.11.  
Questo notebook installa Python 3.11 tramite Conda.

‚è±Ô∏è **Tempo totale:** ~10-15 minuti

```python
# CELLA 1: Verifica versione Python attuale
import sys
print(f"üêç Python attuale: {sys.version}")
print(f"üìç Path: {sys.executable}")
```

```python
# CELLA 2: Installa Miniconda (se non presente)
%%bash
if [ ! -d "/root/miniconda3" ]; then
    echo "üì¶ Installazione Miniconda..."
    wget -q https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh -O miniconda.sh
    bash miniconda.sh -b -p /root/miniconda3
    rm miniconda.sh
    echo "‚úÖ Miniconda installato!"
else
    echo "‚úÖ Miniconda gi√† presente"
fi
```

```python
# CELLA 3: Crea ambiente Python 3.11 e installa TTS
%%bash
export PATH="/root/miniconda3/bin:$PATH"
source /root/miniconda3/etc/profile.d/conda.sh

# Crea ambiente se non esiste
if ! conda env list | grep -q "tts_env"; then
    echo "üîß Creazione ambiente Python 3.11..."
    conda create -n tts_env python=3.11 -y
fi

# Attiva e installa pacchetti
conda activate tts_env
echo "üì¶ Installazione TTS e dipendenze..."
pip install -q TTS==0.22.0 soundfile numpy==1.26.4
echo "‚úÖ Installazione completata!"
python --version
```

```python
# CELLA 4: Test XTTS v2
%%bash
export PATH="/root/miniconda3/bin:$PATH"
source /root/miniconda3/etc/profile.d/conda.sh
conda activate tts_env

python << 'EOF'
import torch
from TTS.api import TTS
import soundfile as sf

print("üéÆ GPU Check:")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"VRAM: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

print("\nüîÑ Caricamento XTTS v2...")
device = "cuda" if torch.cuda.is_available() else "cpu"
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)
print("‚úÖ XTTS v2 caricato!")

print("\nüß™ Generazione audio di test...")
test_text = "Ciao! Questo √® un test di sintesi vocale italiana su Google Colab con Python 3.11."

tts.tts_to_file(
    text=test_text,
    file_path="/content/test_output.wav",
    language="it"
)

print("‚úÖ Test completato!")
print("üìÅ File salvato: /content/test_output.wav")
EOF
```
