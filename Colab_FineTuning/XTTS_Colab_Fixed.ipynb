```python
%%bash
# CELLA 2: Installa Miniconda
if [ ! -d "/root/miniconda3" ]; then
    echo "üì¶ Installazione Miniconda..."
    wget -q https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh -O miniconda.sh
    bash miniconda.sh -b -p /root/miniconda3
    rm miniconda.sh
    echo "‚úÖ Miniconda installato!"
else
    echo "‚úÖ Miniconda gi√† presente"
fi
```

```python
%%bash
# CELLA 3: Setup Conda con conda-forge (NO ToS)
export PATH="/root/miniconda3/bin:$PATH"
source /root/miniconda3/etc/profile.d/conda.sh

echo "üîß Configurazione Conda con conda-forge..."
conda config --remove channels defaults 2>/dev/null || true
conda config --add channels conda-forge
conda config --set channel_priority strict
echo "‚úÖ Conda configurato!"
```

```python
%%bash
# CELLA 4: Crea ambiente Python 3.11
export PATH="/root/miniconda3/bin:$PATH"
source /root/miniconda3/etc/profile.d/conda.sh

# Rimuovi ambiente esistente se presente
conda env remove -n tts_env -y 2>/dev/null || true

echo "üîß Creazione ambiente Python 3.11..."
conda create -n tts_env python=3.11 -y -c conda-forge

echo "‚úÖ Ambiente creato!"
conda env list
```

```python
%%bash
# CELLA 5: Verifica Python 3.11
export PATH="/root/miniconda3/bin:$PATH"
source /root/miniconda3/etc/profile.d/conda.sh
conda activate tts_env

echo "üêç Verifica versione Python:"
python --version
which python
```

```python
%%bash
# CELLA 6: Installa TTS e dipendenze
export PATH="/root/miniconda3/bin:$PATH"
source /root/miniconda3/etc/profile.d/conda.sh
conda activate tts_env

echo "üì¶ Installazione TTS 0.22.0..."
pip install --no-cache-dir TTS==0.22.0

echo "\nüì¶ Installazione dipendenze audio..."
pip install --no-cache-dir soundfile librosa numpy==1.26.4

echo "\n‚úÖ Installazione completata!"
pip show TTS | grep Version
```

```python
%%bash
# CELLA 7: Test XTTS v2
export PATH="/root/miniconda3/bin:$PATH"
source /root/miniconda3/etc/profile.d/conda.sh
conda activate tts_env

python << 'EOF'
import torch
from TTS.api import TTS
import sys

print(f"üêç Python in uso: {sys.version}")
print(f"\nüéÆ GPU Check:")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"VRAM: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

print("\nüîÑ Caricamento XTTS v2...")
device = "cuda" if torch.cuda.is_available() else "cpu"
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)
print("‚úÖ XTTS v2 caricato!")

print("\nüß™ Generazione audio di test...")
test_text = "Ciao! Questo √® un test di sintesi vocale italiana con XTTS versione 2 su Google Colab. Il setup √® finalmente funzionante!"

tts.tts_to_file(
    text=test_text,
    file_path="/content/test_output.wav",
    language="it"
)

print("\n‚úÖ Test completato!")
print("üìÅ File salvato: /content/test_output.wav")
EOF
```

```python
# CELLA 8: Ascolta il risultato
from IPython.display import Audio
print("üîä Audio generato:")
Audio("/content/test_output.wav")
```
