````python
%%bash
# CELLA 2: Installa Miniconda
if [ ! -d "/root/miniconda3" ]; then
    echo "üì¶ Installazione Miniconda..."
    wget -q https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh -O miniconda.sh
    bash miniconda.sh -b -p /root/miniconda3
    rm miniconda.sh
    echo "‚úÖ Miniconda installato!"
else
    echo "‚úÖ Miniconda gi√† presente"
fi
````

````python
%%bash
# CELLA 3: Configura Conda (SOLO conda-forge, NO defaults)
export PATH="/root/miniconda3/bin:$PATH"
eval "$(/root/miniconda3/bin/conda shell.bash hook)"

echo "üîß Rimozione canali defaults..."
conda config --remove channels defaults 2>/dev/null || true
conda config --remove channels pkgs/main 2>/dev/null || true
conda config --remove channels pkgs/r 2>/dev/null || true

echo "üîß Configurazione conda-forge..."
conda config --add channels conda-forge
conda config --set channel_priority strict
conda config --set auto_activate_base false

echo "\n‚úÖ Configurazione completata!"
echo "\nüìã Canali attivi:"
conda config --show channels
````

````python
%%bash
# CELLA 4: Crea ambiente Python 3.11 (FORZATO)
export PATH="/root/miniconda3/bin:$PATH"
eval "$(/root/miniconda3/bin/conda shell.bash hook)"

echo "üóëÔ∏è Pulizia ambiente precedente..."
conda env remove -n tts_env -y 2>/dev/null || true

echo "\nüîß Creazione ambiente Python 3.11 (solo conda-forge)..."
conda create -n tts_env python=3.11 -y --override-channels -c conda-forge

echo "\n‚úÖ Ambiente creato!"
conda env list
````

````python
%%bash
# CELLA 5: VERIFICA CRITICA - Python 3.11
export PATH="/root/miniconda3/bin:$PATH"
eval "$(/root/miniconda3/bin/conda shell.bash hook)"
conda activate tts_env

echo "üêç VERIFICA Python versione:"
python --version
echo "\nüìç Path Python:"
which python
echo "\nüîç Info Python:"
python -c "import sys; print(f'Versione: {sys.version}'); print(f'Executable: {sys.executable}')"
````

````python
%%bash
# CELLA 6: Installa TTS 0.22.0
export PATH="/root/miniconda3/bin:$PATH"
eval "$(/root/miniconda3/bin/conda shell.bash hook)"
conda activate tts_env

echo "üì¶ Installazione TTS 0.22.0..."
pip install --no-cache-dir TTS==0.22.0

echo "\nüì¶ Installazione dipendenze..."
pip install --no-cache-dir soundfile librosa numpy==1.26.4

echo "\n‚úÖ Installazione completata!"
echo "\nüìã Versione TTS installata:"
pip show TTS | grep -E "Name|Version|Requires-Python"
````

````python
%%bash
# CELLA 7: Test XTTS v2
export PATH="/root/miniconda3/bin:$PATH"
eval "$(/root/miniconda3/bin/conda shell.bash hook)"
conda activate tts_env

python << 'PYTHON_SCRIPT'
import sys
import torch
from TTS.api import TTS

print("="*60)
print("üêç AMBIENTE ATTIVO")
print("="*60)
print(f"Python: {sys.version}")
print(f"Executable: {sys.executable}")

print("\n" + "="*60)
print("üéÆ GPU INFO")
print("="*60)
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"VRAM: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

print("\n" + "="*60)
print("üîÑ CARICAMENTO MODELLO XTTS v2")
print("="*60)
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Device: {device}")
print("Download modello in corso...")
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)
print("‚úÖ XTTS v2 caricato!")

print("\n" + "="*60)
print("üß™ GENERAZIONE AUDIO TEST")
print("="*60)
test_text = """Ciao! Questo √® un test di sintesi vocale italiana con XTTS versione 2. 
Il setup su Google Colab √® finalmente funzionante con Python 3.11. 
Possiamo ora procedere con il voice cloning e il fine-tuning!"""

print(f"Testo: {test_text[:50]}...")
print("Generazione in corso...")

tts.tts_to_file(
    text=test_text,
    file_path="/content/test_output.wav",
    language="it"
)

print("\n" + "="*60)
print("‚úÖ TEST COMPLETATO CON SUCCESSO!")
print("="*60)
print("üìÅ File salvato: /content/test_output.wav")

# Verifica file creato
import os
if os.path.exists("/content/test_output.wav"):
    file_size = os.path.getsize("/content/test_output.wav") / 1024
    print(f"üìä Dimensione file: {file_size:.2f} KB")
else:
    print("‚ö†Ô∏è ERRORE: File non trovato!")
PYTHON_SCRIPT
````
